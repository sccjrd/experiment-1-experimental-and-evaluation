```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data from CSV files
df_integer = pd.read_csv('ResultsInteger.csv')
df_string = pd.read_csv('ResultsString.csv')

# Add a column to differentiate between integer and string arrays
df_integer['ArrayType'] = 'Integer'
df_string['ArrayType'] = 'String'

# # Combine the dataframes
# df_integer = pd.concat([df_integer, df_string])

# Ensure the necessary columns exist
required_columns = ['ArraySize', 'Type', 'Iteration', 'Time(ns)', 'Algorithm']
missing_columns = [col for col in required_columns if col not in df_integer.columns]
if missing_columns:
    raise KeyError(f"Missing columns in the combined dataframe: {missing_columns}")

missing_columns = [col for col in required_columns if col not in df_string.columns]
if missing_columns:
    raise KeyError(f"Missing columns in the combined dataframe: {missing_columns}")

# Convert Iteration column to numeric if it's not already
df_integer['Iteration'] = pd.to_numeric(df_integer['Iteration'], errors='coerce')
df_string['Iteration'] = pd.to_numeric(df_string['Iteration'], errors='coerce')

# Function to plot data for a given array size and type
def plot_data(df, array_size, array_type):
    df_filtered = df[(df['ArraySize'] == array_size) & (df['Type'] == array_type)]
    if df_filtered.empty:
        print(f"No data found for ArraySize={array_size} and Type={array_type}")
        return
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=df_filtered, x="Iteration", y="Time(ns)", hue="Algorithm", marker="o")
    plt.title(f'Comparison of Sorting Algorithms for Array Size {array_size} and Type {array_type}')
    plt.xlabel('Iteration')
    plt.ylabel('Time (ns)')
    plt.legend(title='Algorithm')
    plt.show()

print("Integer")    

# Plot data for array size 20 and type "Integer"
plot_data(df_integer, 20, 'random')
plot_data(df_integer, 20, 'ordered')
plot_data(df_integer, 20, 'inverse ordered')


# Plot data for array size 400 and type "Integer"
plot_data(df_integer, 400, 'random')
plot_data(df_integer, 400, 'ordered')
plot_data(df_integer, 400, 'inverse ordered')

# Plot data for array size 10000 and type "Integer"
plot_data(df_integer, 10000, 'random')
plot_data(df_integer, 10000, 'ordered')
plot_data(df_integer, 10000, 'inverse ordered')

print("String")
# Plot data for array size 20 and type "Integer"
plot_data(df_string, 20, 'random')
plot_data(df_string, 20, 'ordered')
plot_data(df_string, 20, 'inverse ordered')

# Plot data for array size 400 and type "Integer"
plot_data(df_string, 400, 'random')
plot_data(df_string, 400, 'ordered')
plot_data(df_string, 400, 'inverse ordered')

# Plot data for array size 10000 and type "Integer"
plot_data(df_string, 10000, 'random')
plot_data(df_string, 10000, 'ordered')
plot_data(df_string, 10000, 'inverse ordered')

warmup_iterations = 100

# Compute the average time for array size 20 and type "random" from iteration number warmup_iterations onwards
df_filtered = df_integer[(df_integer['ArraySize'] == 20) & (df_integer['Type'] == 'random') & (df_integer['Iteration'] >= warmup_iterations)]
average_time = df_filtered['Time(ns)'].mean()
print(f"The average time for array size 20 and type 'random' from iteration number {warmup_iterations} onwards is {average_time} ns")

# Compute the average time for array size 20 and type "ordered" from iteration number warmup_iterations onwards
df_filtered = df_integer[(df_integer['ArraySize'] == 20) & (df_integer['Type'] == 'ordered') & (df_integer['Iteration'] >= warmup_iterations)]
average_time = df_filtered['Time(ns)'].mean()
print(f"The average time for array size 20 and type 'ordered' from iteration number {warmup_iterations} onwards is {average_time} ns")

# Compute the average time for array size 20 and type "inverse ordered" from iteration number warmup_iterations onwards
df_filtered = df_integer[(df_integer['ArraySize'] == 20) & (df_integer['Type'] == 'inverse ordered') & (df_integer['Iteration'] >= warmup_iterations)]
average_time = df_filtered['Time(ns)'].mean()
print(f"The average time for array size 20 and type 'inverse ordered' from iteration number {warmup_iterations} onwards is {average_time} ns")

warmup_iterations = 30

# Compute the average time for array size 400 and type "random" from iteration number warmup_iterations onwards
df_filtered = df_integer[(df_integer['ArraySize'] == 400) & (df_integer['Type'] == 'random') & (df_integer['Iteration'] >= warmup_iterations)]
average_time = df_filtered['Time(ns)'].mean()
print(f"The average time for array size 400 and type 'random' from iteration number {warmup_iterations} onwards is {average_time} ns")

# Compute the average time for array size 400 and type "ordered" from iteration number warmup_iterations onwards
df_filtered = df_integer[(df_integer['ArraySize'] == 400) & (df_integer['Type'] == 'ordered') & (df_integer['Iteration'] >= warmup_iterations)]
average_time = df_filtered['Time(ns)'].mean()
print(f"The average time for array size 400 and type 'ordered' from iteration number {warmup_iterations} onwards is {average_time} ns")

# Compute the average time for array size 400 and type "inverse ordered" from iteration number warmup_iterations onwards
df_filtered = df_integer[(df_integer['ArraySize'] == 400) & (df_integer['Type'] == 'inverse ordered') & (df_integer['Iteration'] >= warmup_iterations)]
average_time = df_filtered['Time(ns)'].mean()
print(f"The average time for array size 400 and type 'inverse ordered' from iteration number {warmup_iterations} onwards is {average_time} ns")

warmup_iterations = 75

# Compute the average time for array size 10000 and type "random" from iteration number warmup_iterations onwards
df_filtered = df_integer[(df_integer['ArraySize'] == 10000) & (df_integer['Type'] == 'random') & (df_integer['Iteration'] >= warmup_iterations)]
average_time = df_filtered['Time(ns)'].mean()
print(f"The average time for array size 10000 and type 'random' from iteration number {warmup_iterations} onwards is {average_time} ns")

# Compute the average time for array size 10000 and type "ordered" from iteration number warmup_iterations onwards
df_filtered = df_integer[(df_integer['ArraySize'] == 10000) & (df_integer['Type'] == 'ordered') & (df_integer['Iteration'] >= warmup_iterations)]
average_time = df_filtered['Time(ns)'].mean()
print(f"The average time for array size 10000 and type 'ordered' from iteration number {warmup_iterations} onwards is {average_time} ns")

# Compute the average time for array size 10000 and type "inverse ordered" from iteration number warmup_iterations onwards
df_filtered = df_integer[(df_integer['ArraySize'] == 10000) & (df_integer['Type'] == 'inverse ordered') & (df_integer['Iteration'] >= warmup_iterations)]
average_time = df_filtered['Time(ns)'].mean()
print(f"The average time for array size 10000 and type 'inverse ordered' from iteration number {warmup_iterations} onwards is {average_time} ns")

# Repeat the same for df_string

warmup_iterations = 100

# Compute the average time for array size 20 and type "random" from iteration number warmup_iterations onwards
df_filtered = df_string[(df_string['ArraySize'] == 20) & (df_string['Type'] == 'random') & (df_string['Iteration'] >= warmup_iterations)]
average_time = df_filtered['Time(ns)'].mean()
print(f"The average time for array size 20 and type 'random' from iteration number {warmup_iterations} onwards is {average_time} ns")

# Compute the average time for array size 20 and type "ordered" from iteration number warmup_iterations onwards
df_filtered = df_string[(df_string['ArraySize'] == 20) & (df_string['Type'] == 'ordered') & (df_string['Iteration'] >= warmup_iterations)]
average_time = df_filtered['Time(ns)'].mean()
print(f"The average time for array size 20 and type 'ordered' from iteration number {warmup_iterations} onwards is {average_time} ns")

# Compute the average time for array size 20 and type "inverse ordered" from iteration number warmup_iterations onwards
df_filtered = df_string[(df_string['ArraySize'] == 20) & (df_string['Type'] == 'inverse ordered') & (df_string['Iteration'] >= warmup_iterations)]
average_time = df_filtered['Time(ns)'].mean()
print(f"The average time for array size 20 and type 'inverse ordered' from iteration number {warmup_iterations} onwards is {average_time} ns")

warmup_iterations = 30

# Compute the average time for array size 400 and type "random" from iteration number warmup_iterations onwards
df_filtered = df_string[(df_string['ArraySize'] == 400) & (df_string['Type'] == 'random') & (df_string['Iteration'] >= warmup_iterations)]
average_time = df_filtered['Time(ns)'].mean()
print(f"The average time for array size 400 and type 'random' from iteration number {warmup_iterations} onwards is {average_time} ns")

# Compute the average time for array size 400 and type "ordered" from iteration number warmup_iterations onwards
df_filtered = df_string[(df_string['ArraySize'] == 400) & (df_string['Type'] == 'ordered') & (df_string['Iteration'] >= warmup_iterations)]
average_time = df_filtered['Time(ns)'].mean()
print(f"The average time for array size 400 and type 'ordered' from iteration number {warmup_iterations} onwards is {average_time} ns")

# Compute the average time for array size 400 and type "inverse ordered" from iteration number warmup_iterations onwards
df_filtered = df_string[(df_string['ArraySize'] == 400) & (df_string['Type'] == 'inverse ordered') & (df_string['Iteration'] >= warmup_iterations)]
average_time = df_filtered['Time(ns)'].mean()
print(f"The average time for array size 400 and type 'inverse ordered' from iteration number {warmup_iterations} onwards is {average_time} ns")

warmup_iterations = 75

# Compute the average time for array size 10000 and type "random" from iteration number warmup_iterations onwards
df_filtered = df_string[(df_string['ArraySize'] == 10000) & (df_string['Type'] == 'random') & (df_string['Iteration'] >= warmup_iterations)]
average_time = df_filtered['Time(ns)'].mean()
print(f"The average time for array size 10000 and type 'random' from iteration number {warmup_iterations} onwards is {average_time} ns")

# Compute the average time for array size 10000 and type "ordered" from iteration number warmup_iterations onwards
df_filtered = df_string[(df_string['ArraySize'] == 10000) & (df_string['Type'] == 'ordered') & (df_string['Iteration'] >= warmup_iterations)]
average_time = df_filtered['Time(ns)'].mean()
print(f"The average time for array size 10000 and type 'ordered' from iteration number {warmup_iterations} onwards is {average_time} ns")

# Compute the average time for array size 10000 and type "inverse ordered" from iteration number warmup_iterations onwards
df_filtered = df_string[(df_string['ArraySize'] == 10000) & (df_string['Type'] == 'inverse ordered') & (df_string['Iteration'] >= warmup_iterations)]
average_time = df_filtered['Time(ns)'].mean()
print(f"The average time for array size 10000 and type 'inverse ordered' from iteration number {warmup_iterations} onwards is {average_time} ns")


# Find the best algorithm for each array size in the random case for strings and integers
best_algorithms = {}

for array_size in df_integer['ArraySize'].unique():
    df_filtered_integer = df_integer[(df_integer['ArraySize'] == array_size) & (df_integer['Type'] == 'random') & (df_integer['Iteration'] >= warmup_iterations)]
    best_algorithm_integer = df_filtered_integer.groupby('Algorithm')['Time(ns)'].mean().idxmin()
    best_time_integer = df_filtered_integer.groupby('Algorithm')['Time(ns)'].mean().min()
    
    df_filtered_string = df_string[(df_string['ArraySize'] == array_size) & (df_string['Type'] == 'random') & (df_string['Iteration'] >= warmup_iterations)]
    best_algorithm_string = df_filtered_string.groupby('Algorithm')['Time(ns)'].mean().idxmin()
    best_time_string = df_filtered_string.groupby('Algorithm')['Time(ns)'].mean().min()
    
    best_algorithms[array_size] = {
        'Integer': {'Algorithm': best_algorithm_integer, 'Time(ns)': best_time_integer},
        'String': {'Algorithm': best_algorithm_string, 'Time(ns)': best_time_string}
    }

# Print the best algorithms for each array size
for array_size, algorithms in best_algorithms.items():
    print(f"Array Size: {array_size}")
    print(f"  Best Integer Algorithm: {algorithms['Integer']['Algorithm']} with Time(ns): {algorithms['Integer']['Time(ns)']}")
    print(f"  Best String Algorithm: {algorithms['String']['Algorithm']} with Time(ns): {algorithms['String']['Time(ns)']}")


    # Compute the standard deviation for each array size and type
    for array_size in df_integer['ArraySize'].unique():
        df_filtered_integer = df_integer[(df_integer['ArraySize'] == array_size) & (df_integer['Type'] == 'random') & (df_integer['Iteration'] >= warmup_iterations)]
        std_dev_integer = df_filtered_integer.groupby('Algorithm')['Time(ns)'].std()
        
        df_filtered_string = df_string[(df_string['ArraySize'] == array_size) & (df_string['Type'] == 'random') & (df_string['Iteration'] >= warmup_iterations)]
        std_dev_string = df_filtered_string.groupby('Algorithm')['Time(ns)'].std()
        
        print(f"Array Size: {array_size}")
        print(f"  Integer Algorithms Standard Deviation:\n{std_dev_integer}")
        print(f"  String Algorithms Standard Deviation:\n{std_dev_string}")
```
